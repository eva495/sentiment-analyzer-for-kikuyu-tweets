# use human anotators to label a sample of the data

```python
# Define the sample text
text = "Hoya Ngai, hoya nama. Hoya nƒ´≈´k≈´heo. Hoya ≈´teg≈´thangania wƒ´theretie."

# Define the possible labels or categories
labels = ["positive", "negative", "neutral"]

# Define a list to store the labeled text
labeled_text = []

# Loop over each sentence in the text and ask the human annotators to label it
for sentence in text.split("."):
    # Print the sentence to be labeled
    print(f"Sentence to label: {sentence}")
    # Ask the human annotators to label the sentence
    label = input("Please label the sentence (positive/negative/neutral): ")
    # Check if the label is valid
    if label not in labels:
        print("Invalid label. Please try again.")
        continue
    # Store the labeled sentence
    labeled_sentence = {"text": sentence, "label": label}
    labeled_text.append(labeled_sentence)

# Print the labeled text
print("Labeled text:")
for sentence in labeled_text:
    print(f"{sentence['text']} ({sentence['label']})")
```

```python
#removing stopwords and tokenizing
import nltk

# Load the Kikuyu stopwords
with open('kikuyu_stopwords.txt', 'r') as f:
    stopwords = set(f.read().split())

# Load the text to tokenize
with open('kikuyu_tweets.txt', encoding='utf-8') as f:
    data = f.read()

# Tokenize the text into individual words
tokens = nltk.word_tokenize(data)

# Remove stop words and punctuation
clean_tokens = [token.lower() for token in tokens if token.lower() not in stopwords and token.isalpha()]

# Print the cleaned tokens
print(clean_tokens)
```

```python
#removing english tweets 
from langdetect import detect
import pandas as pd

# Open the input file and read the data
with open('kikuyu_tweets_cleaned.csv', encoding='utf-8') as file:
    data = file.read()

# Remove URLs
data = re.sub(r'http\S+', '', data)

# Tokenize the text into individual tweets
tweets = data.split('\n')

# Create a list to store the non-English tweets
non_english_tweets = []

# Loop through the tweets and detect the language
for tweet in tweets:
    try:
        lang = detect(tweet)
        if lang != 'en':
            non_english_tweets.append(tweet)
    except:
        pass

# Convert the list of non-English tweets into a Pandas DataFrame
df = pd.DataFrame({'Tweet': non_english_tweets})

# Save the DataFrame as a CSV file
df.to_csv('non_english_tweets.csv', index=False)
```

# removes numbers for the tweets and put the data into a paragraph


```python
import pandas as pd

# Load the non-english tweets from the CSV file
df = pd.read_csv('non_english_tweets.csv')

# Remove the tweets where the user was mentioned with '@'
df = df[~df['Tweet'].str.contains('@[\w]*')]

# Remove the numbering
df['text'] = df['Tweet'].str.replace(r'^\d+\s', '')

# Combine all tweets into a single paragraph
paragraph = ' '.join(df['text'].tolist())

# Print the paragraph
print(paragraph)
```
